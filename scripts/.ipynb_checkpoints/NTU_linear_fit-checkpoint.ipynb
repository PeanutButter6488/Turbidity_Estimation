```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import glob

import sys  
sys.path.insert(0, '../dip_utils')

from matrix_utils import arr_info
from vis_utils import (vis_rgb_cube,
                       vis_hists,
                       vis_pair)
```

```python
# calculate the average of 10 images to reduce noise, and plot the histogram with respect to clear water
img_lib = []
clear = glob.glob('../assets/clear/*.jpg')
NTU1_65 = glob.glob('../assets/NTU1_65/*.jpg')
NTU14_4 = glob.glob('../assets/NTU14_4/*.jpg')
NTU2_45 = glob.glob('../assets/NTU2_45/*.jpg')
NTU5_17 = glob.glob('../assets/NTU5_17/*.jpg')
NTU42_6 = glob.glob('../assets/NTU42_6/*.jpg')
img_lib.append(clear)
img_lib.append(NTU1_65)
img_lib.append(NTU14_4)
img_lib.append(NTU2_45)
img_lib.append(NTU5_17)
img_lib.append(NTU42_6)
```

```python
#Transfer to single greyscale image
def to_single_channel(I):
    if (len(I.shape) > 2):
        Ig = 0.2989 * I[..., 0] + 0.5870 * I[..., 1] + 0.1140 * I[..., 2]
        I = Ig

        I = I/I.ravel().max()
        I = 255*I
        I = np.round(I)
        I[I>255] = 255
        I = I.copy().astype('uint8')
        return I
```

```python
def polygon(I):
    left = I[600:1200, 300:500]
    right = I[600:1200, 720:920]
    return left, right
```

```python
# left_arr = np.sum(left, axis=0)
```

```python
# right_arr = np.sum(right, axis=0)
```

```python
def compute_avg(lis):
    left = []
    right = []
    for img in lis:
        img = plt.imread(img)
        img = to_single_channel(img)
        l, r = polygon(img)
        left.append(l)
        right.append(r)
    l_mean = np.mean([np.mean(arr) for arr in left])
    r_mean = np.mean([np.mean(arr) for arr in right])
    return l_mean, r_mean
```

```python
def compute_diff(lis):
    diff = []
    for img in lis:
        img = plt.imread(img)
        img = to_single_channel(img)
        l, r = polygon(img)
        diff.append(l - r)
    return np.array(diff)
```

```python
avg_left = []
avg_right = []
for i in range(len(img_lib)):
    lm, rm = compute_avg(img_lib[i])
    avg_left.append(lm)
    avg_right.append(rm)
```

```python
diff = []
for i in range(len(img_lib)):
    difference = compute_diff(img_lib[i])
    diff.append(difference)
```

```python
avg_left, avg_right
```

```python
x_value = np.array([0, 1.65, 14.4, 2.45, 5.17, 42.6])
```

```python
theta = np.polyfit(x_value, avg_right, 1)
```

```python
plt.figure()
plt.plot(x_value, avg_right, 'yo')
plt.ylabel('average pixel value')
plt.xlabel('NTU')
plt.plot(x_value, theta[0]*x_value + theta[1])
plt.plot(x_value, avg_left, 'bo')
plt.title('Relationship without Calibration')
```

```python
# Due to different conditions (lighting, angel, etc) when picture were taken, 
# need to calibrate based on difference of left and right clear water sample
```

```python
diff = []
for i in range(len(avg_left)):
    difference = avg_right[i] - avg_left[i]
    diff.append(difference)
```

```python
diff = np.array(diff)
```

```python
cali = abs(diff[0])
```

```python
diff
```

```python
diff += cali
```

```python
theta = np.polyfit(x_value, diff, 1)
```

```python
plt.figure()
plt.plot(x_value, diff, 'yo')
plt.ylabel('calibrated intensity')
plt.xlabel('NTU')
plt.plot(x_value, theta[0]*x_value + theta[1])
plt.title('Calibrated based on Difference')
```

```python
cali2 = avg_right[0] / avg_left[0]
cali2
```

```python
ratio = []
for i in range(len(avg_left)):
    rate = (avg_right[i] / avg_left[i]) / cali2
    ratio.append(rate)
```

```python
ratio
```

```python
theta = np.polyfit(x_value, ratio, 1)
```

```python
plt.figure()
plt.plot(x_value, ratio, 'yo')
plt.ylabel('calibrated intensity ratio')
plt.xlabel('NTU')
plt.plot(x_value, theta[0]*x_value + theta[1])
plt.title('Calibrated based on Ratio')
```
