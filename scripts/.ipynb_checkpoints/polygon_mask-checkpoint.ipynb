```python
%matplotlib widget
from skimage.draw import polygon, polygon2mask
import matplotlib.pyplot as plt
import numpy as np

import sys  
sys.path.insert(0, '../dip_utils')

from matrix_utils import arr_info
from vis_utils import (vis_rgb_cube,
                       vis_hists,
                       vis_pair)
```

```python
# Define masks with 20x20 subgrids, focusing on Red Bar first
coordR = np.array([[680, 400], [680, 420], [700, 420], [700, 400]])
coordR1 = np.array([[680, 700], [680, 720], [700, 720], [700, 700]])
```

```python
coordR[:,0], coordR1[:,0]
```

```python
rrR, ccR = polygon(coordR[:,0], coordR[:,1], shape=I.shape[:2])
rrR1, ccR1 = polygon(coordR1[:,0], coordR1[:,1], shape=I.shape[:2])
```

```python
maskR = np.zeros(I.shape[:2])
maskR[rrR,ccR] = 1
```

```python
maskR1 = np.zeros(I.shape[:2])
maskR1[rrR1,ccR1] = 1
```

```python
arr_info(maskR), arr_info(maskR1)
```

```python
vis_pair(I, np.clip(I*maskR[..., None], 0, 1))
```

```python
I[maskR==1, :].mean(axis=0)
```

```python
I[maskR1==1, :].mean(axis=0)
```
