```python
img_lib = []
```

```python
for dic in os.listdir(path):
    if dic != '.ipynb_checkpoints':
        img_lib.append(dic)
img_lib
```

```python
markers_lib = dict()
for img in img_lib:
    markers = watershed(img)
    markers_lib[img] = markers
markers_lib
```

```python
bottom_lib = dict()
left_lib = dict()
```

```python
# Acquire the bottom and the leftmost point of the ROI
for key in markers_lib:
    bottom = np.where(markers_lib[key]==2)[0].max()
    bottom_lib[key] = bottom
    left = np.where(markers_lib[key]==2)[1].min()
    left_lib[key] = left
bottom_lib, left_lib
```

```python
# Some threshold values
vert_thres = 650
horz_thres = 50
length = 100
dist = 380
```

```python
# Avoid Extracting incorrect ROI
for key in markers_lib:
    while markers_lib[key][bottom_lib[key]-vert_thres, left_lib[key]+horz_thres] == 1:
        left_lib[key] += horz_thres
left_lib
```

```python
# Calibration
calibrated = {}
for img in img_lib:
    org = plt.imread(os.path.join(path, img))
    org = lab_trans(org)
    calibrated[img] = org
```

```python
greyscale = {}
for img in img_lib:
    grey = calibrated[img]
    grey = to_single_channel(grey)
    greyscale[img] = grey
greyscale
```

```python
# x_left represents the ROI of the reference, while x_right represents the ROI of the sample
x_left = []
x_right = []
y_left = []
y_right = []
```

```python
for key in greyscale:
    x = []
    x1 = []
    y = []
    y1 = []
    for i in range(bottom_lib[key], bottom_lib[key]-vert_thres, -1):
        x.append(i)
        y.append(np.mean(greyscale[key][i-1:i, left_lib[key]+horz_thres:left_lib[key]+horz_thres+length]))
    x_left.append(x)
    y_left.append(y)
    for i in range(bottom_lib[key], bottom_lib[key]-vert_thres, -1):
        x1.append(i)
        y1.append(np.mean(greyscale[key][i-1:i, left_lib[key]+horz_thres+dist:left_lib[key]+horz_thres+dist+length]))
    x_right.append(x1)
    y_right.append(y1)
```

```python
x_left = np.array(x_left)
x_right = np.array(x_right)
y_left = np.array(y_left)
y_right = np.array(y_right)
```

```python
arr_info(x_left), arr_info(y_left)
```

```python
theta_l = {}
theta_r = {}
for i in range(x_left.shape[0]):
    tl = np.polyfit(x_left[i], y_left[i], 1)
    tr = np.polyfit(x_right[i], y_right[i], 1)
    theta_l[img_lib[i]] = tl
    theta_r[img_lib[i]] = tr
```

```python
theta_l, theta_r
```

```python
fig, ax = plt.subplots(1,x_left.shape[0], figsize=(15, 5))
for i in range(x_left.shape[0]):
    ax[i].plot(x_left[i], y_left[i], label='%s ref'%img_lib[i])
    ax[i].plot(x_right[i], y_right[i], label='%s samp'%img_lib[i])
    ax[i].plot(x_left[i], theta_l[img_lib[i]][0]*x_left[i]+theta_l[img_lib[i]][1], label='regression ref')
    ax[i].plot(x_right[i], theta_r[img_lib[i]][0]*x_right[i]+theta_r[img_lib[i]][1], label='regression samp')
    ax[i].legend()
    ax[i].set_title("Mean Value from Bottom to Top")
```
