```python
def convert_temp(I):
    R = np.mean(I[..., 0])
    G = np.mean(I[..., 1])
    B = np.mean(I[..., 2])
    X = (-0.14282)*(R) + (1.54924)*(G) + (-0.95641)*(B)
    Y = (-0.32466)*(R) + (1.57837)*(G) + (-0.73191)*(B)  # illuminance
    Z = (-0.68202)*(R) + (0.77073)*(G) + (0.56332)*(B)
    x = X/(X+Y+Z) 
    y = Y/(X+Y+Z)
    n = (x - 0.3320) / (0.1858 - y)
    CCT = 449*(n**3) + 3525*(n**2) + 6823.3*(n) + 5520.33
    return np.mean(CCT)
```
