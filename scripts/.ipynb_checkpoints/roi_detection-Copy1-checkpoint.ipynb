```python
%matplotlib widget
import numpy as np
from matplotlib import pyplot as plt
from skimage.draw import polygon, polygon2mask
import glob
import numpy as np
import cv2
from matplotlib import pyplot as plt

import os
import sys  
sys.path.insert(0, '../dip_utils')

from matrix_utils import arr_info
from vis_utils import (vis_rgb_cube,
                       vis_hists,
                       vis_pair, 
                       vis_triple)
%run calibrate.ipynb
%run greyscale.ipynb
```

```python
filename = 'NTU0'
path = '../stacked_new/NTU0.jpg'
img = plt.imread(path)
```

```python
# Some threshold values
bottom = 1480
top = 1000
left = 420
length = 50
dist = 350
right = 750
```

```python
#If it's a multi-channel image, try to get it in [0,255]
def to_single_channel(I):
    if (len(I.shape) > 2):
        Ig = 0.2989 * I[..., 0] + 0.5870 * I[..., 1] + 0.1140 * I[..., 2]
        I = Ig

        I = I/I.ravel().max()
        I = 255*I
        I = np.round(I)
        I[I>255] = 255
        I = I.copy().astype('uint8')
        return I
```

```python
img = lab_trans(img)
img = to_single_channel(img)
arr_info(img)
plt.figure()
plt.imshow(img, cmap='gray')
```

```python
x = []
x1 = []
y = []
y1 = []
for i in range(bottom, top, -1):
    x.append(i)
    y.append(np.mean(img[i-1:i, left:left+length]))
for i in range(bottom, top, -1):
    x1.append(i)
    y1.append(np.mean(img[i-1:i, right:right+length]))
```

```python
x = np.array(x)
x1 = np.array(x1)
y = np.array(y)
y1 = np.array(y1)
```

```python
plt.figure()
plt.plot(x,y,label='ref')
plt.plot(x1,y1,label='sample')
plt.title("Mean Value From Bottom to Top %s" % filename)
plt.legend()
plt.show()
```

```python
n = np.polyfit(x, y, 1)
n
n1 = np.polyfit(x1, y1, 1)
n, n1
```

```python
r = n1[0] / n[0]
diff = np.mean(y - y1)
r, diff
```

```python
# y *= r
y1 += diff
```

```python
plt.figure()
plt.plot(x,y,label='ref')
plt.plot(x1,y1,label='sample')
plt.title("Mean Value From Bottom to Top %s" % filename)
plt.legend()
plt.show()
```

```python
NTU = [0, 99.4, 198, 445]
y_value = []
y_value.append(np.mean(y1))
```
