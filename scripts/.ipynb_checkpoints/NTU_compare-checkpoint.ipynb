```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

import sys  
sys.path.insert(0, '../dip_utils')

from matrix_utils import arr_info
from vis_utils import (vis_rgb_cube,
                       vis_hists,
                       vis_pair)
```

```python
clear_black = plt.imread('../assets/clear_black.jpg')
```

```python
plt.figure()
plt.imshow(clear_black)
```

```python
#If it's a multi-channel image, try to get it in [0,255]
def to_single_channel(I):
    if (len(I.shape) > 2):
        Ig = 0.2989 * I[..., 0] + 0.5870 * I[..., 1] + 0.1140 * I[..., 2]
        I = Ig

        I = I/I.ravel().max()
        I = 255*I
        I = np.round(I)
        I[I>255] = 255
        I = I.copy().astype('uint8')
        return I
```

```python
clear_black = to_single_channel(clear_black)
```

```python
clear_black_left = clear_black[240:1240, 220:500]
clear_black_right = clear_black[240:1240, 740:1020]
```

```python
histcbl, binscbl = np.histogram(clear_black_left.ravel(), np.arange(257))
histcbr, binscbr = np.histogram(clear_black_right.ravel(), np.arange(257))
```

```python
f, ax = plt.subplots(1,2, figsize=(8,3), sharex=True, sharey=True)

ax[0].bar(binscbl[:-1], histcbl)
ax[0].set_title('CWSL with Black Background')

ax[1].bar(binscbr[:-1], histcbr)
ax[1].set_title('CWSR with Black Background')
```

```python
np.mean(clear_black_left), np.mean(clear_black_right)
```
